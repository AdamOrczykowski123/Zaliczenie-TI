```python
import numpy as np
import matplotlib.pyplot as plt
Uc = 6.0
R = 6250
C1 = 0.00047
C2 = 0.00094
e = 2,718281828459045235360
t = t = np.linspace(0, 10, 1000)
U1 = Uc * e**(-t/(R*C1))
U2 = Uc * e**(-t/(R*C2))
plt.grid()
plt.xlabel ("Czas")
plt.ylabel ("Napięcie")
plt.title("Wykres ładowania kondensatora")
plt.legend()
plt.plot(t, U1)
plt.plot(t, U2)
plt.show()
plt.figure()
```
